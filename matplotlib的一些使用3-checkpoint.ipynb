```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-np.pi,np.pi,500)
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
ax=plt.gca()
ax.spines['right'].set_color('none')  # 隐藏右侧的边框  注意：【】 而不是（）
ax.spines['top'].set_color('none')
```

```python
plt.plot(x,y)
ax=plt.gca()
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
```

```python
#plt.figure()
plt.plot(x,y)
ax=plt.gca()
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$1$'])
#plt.show()
```

```python
x=np.random.normal(20,3,10000)
#plt.plot(x)
#plt.scatter([x for x in range(len(x))],x)
plt.subplot(121)
plt.hist(x,bins=60,normed=True) # normed =True 表示归一化（即显示为概率密度）
ax=plt.gca()
ax.set_xlabel('vaule')
ax.set_ylabel('possibility')
ax.set_title(r'$\mathrm{Histogram:}\ \mu=%d,\ \sigma=%d$' %(20,3)) # 如何设置标题的显示，字体之类的 $ $ 之间的是公式 \\ 之间是部分
plt.subplot(122)
plt.hist(x,bins=60) 
ax=plt.gca()
ax.set_xlabel('vaule')
ax.yaxis.set_ticks_position('right')
ax.set_ylabel('acount')
ax.yaxis.set_label_position('right')                              # 如同设置ticks的位置一样
ax.set_title(r'$\mathrm{Histogram:}\ \mu=%d,\ \sigma=%d$' %(20,3)) 
```
