```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# x,y的取值范围
x=np.linspace(-5,5,201)
y=np.linspace(-5,5,201)
```

```python
x
```

```python
y
```

```python
# 高斯分布
z=np.exp((-0.5)*((x*x)+(y*y))/2)
```

```python
fig=plt.figure(figsize=(6,6))
ax3=plt.axes(projection='3d')
ax3.plot(x,y,z)
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Z')  
ax3.view_init(elev=20,azim=135)
plt.show()
```

可以看到，当仅仅是定义了x,y的取值范围，并用来计算高斯分布的时候，我们得到的是一条曲线，而不是一个曲面。这是因为在x，y的这种定义下，转换到坐标系中是这样的：

也就是整个平面上的一条对角线。  
如果想要画出曲面来， 需要把这个平面上所有的点都取到，应如下图所示：

而np.meshgrid就是来做这件事的

```python
X,Y=np.meshgrid(x,y)
X
# x1,y1=range(5),range(5,10)
# X1,Y1=np.meshgrid(x1,y1)
# X1
```

可以看到，对于X，每一行对应的就是x；而X的行数跟y的取点个数一样

```python
Y
# Y1
```

可以看到，对于Y，每一列都是y，而Y的列数跟x的取点个数一样

```python
# 计算高斯分布
Z=np.exp((-0.5)*((X*X)+(Y*Y))/2)
```

```python
# 画出曲面
fig=plt.figure(figsize=(6,6))
ax3=plt.axes(projection='3d')
ax3.plot_surface(X,Y,Z,cmap='rainbow',rstride=1,cstride=1)
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Z')    

# 设定观察视角，elev绕水平平行于纸面的轴转动角度；azim绕z轴转动角度
ax3.view_init(elev=20,azim=135)
plt.show()
```

```python
# 画出曲面和投影到不同平面的等高线
fig=plt.figure(figsize=(6,6))
ax3=plt.axes(projection='3d')

# cstride，rstride，图片精细度，越小越精细，最小为1
ax3.plot_surface(X,Y,Z,cmap='rainbow',rstride=1,cstride=1)
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Z')

# 画出投影到不同平面的等高线
# zdir=‘z’.投影到z平面，offset=0，画在z=0的平面上
ax3.contour(X,Y,Z,5,zdir='z',offset=0,cmap='rainbow')   

# zdir=‘x’.投影到x平面，offset=5，画在x=5的平面上
ax3.contour(X,Y,Z,zdir='x',offset=5,)    
ax3.contour(X,Y,Z,zdir='y',offset=-6,)    

# view_init 设定图片观察视角，elev绕水平平行于纸面的轴转动角度；azim绕z轴转动角度
ax3.view_init(elev=20,azim=135)
plt.show()
```

```python
# 使用热图的形式画出，即使用颜色的值代表数据格点上的值
plt.figure(figsize=(6,5))
plt.pcolormesh(X,Y,Z,shading='auto')
# plt.pcolormesh(X,Y,Z)
plt.colorbar()
plt.axis('equal')
plt.show()
```

接下来看一下，如果自己的实验/模拟数据是矩阵形式的（相当于数据点铺满一个平面的）

```python
# 文件地址
# 需要注意斜杠的方向
path='数据所在盘/文件夹/文件名.csv'
skip_rows=8
skip_cols=1
# x 对应的是时间，也就是csv文件中的列
xmin=0
xmax=80

# y对应的是深度，也就是csv文件中的行
ymin=0
ymax=2
level_value=2300
```

```python
data=pd.read_csv(path,skiprows=skip_rows,header=None)
data=data.iloc[:,skip_cols:]
data.head()
```

```python
# 定义读取数据函数
def read_data(path=path,level_value=level_value,x_min=xmin,x_max=xmax,y_min=ymin,y_max=ymax,skip_rows=skip_rows,skip_cols=skip_cols):
    data=pd.read_csv(str(path),skiprows=skip_rows,header=None)
    data=data.iloc[:,skip_cols:]
    xd=data.shape[1]
    yd=data.shape[0]
    # 构造跟data等大的X和Y，提供坐标
    # 注意：x是csv文件对应的列
    # 所以这里x要用列，即shape[1]
    # y同理
    x=np.linspace(x_min,x_max,xd)
    y=np.linspace(y_min,y_max,yd)
    X,Y=np.meshgrid(x,y)
    return ([X,Y,data])
data=read_data()
```

```python
# 定义画3维图函数
def plt_3d(data):
    X=data[0]
    Y=data[1]
    Z=data[2]
    fig=plt.figure()
    ax3=plt.axes(projection='3d')
    ax3.plot_surface(X,Y,Z,alpha=1,cmap='rainbow',rstride=30,cstride=20)
    ax3.view_init(elev=20,azim=135)
    plt.show()
```

```python
plt_3d(data)
```

```python

# 画出热图形式
X,Y,Z=data
plt.pcolormesh(X,Y,Z,shading='auto',cmap='rainbow')
plt.colorbar()
plt.show()
```

```python
# 画出等高线
# X,Y，Z之后的参数如果是正整数，则表示要画出的等高线的条数
plt.contour(X,Y,Z,10)
plt.show()
```

```python
# X,Y,Z之后的参数如果是列表，比如[2000],则表示画出数值为2000的等高线
plt.contour(X,Y,Z,[2000])
plt.show()
```

```python
# 提出等高线的数据
cl=plt.contour(X,Y,Z,[2000]).collections[0].get_paths()
dn=cl[0].vertices
```

```python
# 重新画出等高线11
plt.plot(dn[:,0],dn[:,1])
plt.show()
```
